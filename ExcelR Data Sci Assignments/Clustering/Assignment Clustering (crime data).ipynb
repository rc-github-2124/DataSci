```python
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering as AC
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm=norm_func(df.iloc[:,1:])
```

```python
df_norm
```

```python
cor = df_norm.corr() #Calculate the correlation of the above variables
sns.heatmap(cor, square = True) 
```

```python
sns.distplot(df_norm)
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
model=AC(n_clusters=4,affinity='euclidean',linkage='complete').fit(df_norm)
```

```python
clusters_labels=pd.Series(model.labels_)
```

```python
df['Clusters']=clusters_labels
```

```python
df=df.iloc[:,[5,0,1,2,3,4]]
```
